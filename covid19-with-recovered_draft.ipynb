```
# importing the required libraries
import pandas as pd
import numpy as np

# Visualisation libraries
import plotly.express as px
import plotly.graph_objects as go
```

```
df_data= pd.read_csv('https://github.com/datasets/covid-19/raw/master/data/time-series-19-covid-combined.csv')
df = df_data.copy()
df = df.rename(columns={'Country/Region': 'Country'})
# drop country/Region column
df.drop('Province/State', inplace=True, axis=1)
# NaN values and negative values replace with 0
df['New Confirmed'] = df.Confirmed.diff().fillna(0).apply(lambda x : x if x > 0 else 0)
df['New Recovered'] = df.Recovered.diff().fillna(0).apply(lambda x : x if x > 0 else 0)
df['New Deaths'] = df.Deaths.diff().fillna(0).apply(lambda x : x if x > 0 else 0)
df
```

```
df_bd = df.query('Country == "Bangladesh" and Confirmed > 0')
# df_bd.to_csv(r'Bangladesh-cases.csv', index = False) # download
df_bd
```

```
fig = px.bar(df_bd, x="Date", y="New Confirmed", barmode='group', height=600)
fig.update_layout(title_text='Coronavirus Confirmed in Bangladesh', plot_bgcolor='rgb(230, 230, 230)')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_bd['Date'], y = df_bd['New Confirmed'], mode='lines+markers',name='Total Cases'))
fig.update_layout(title_text='Coronavirus Confirmed in Bangladesh', plot_bgcolor='rgb(230, 230, 230)')
fig.show()
```

```
# Confirmed vs Recovered
fig = go.Figure(data=[
    go.Bar(x=df_bd["Date"], y=df_bd["New Recovered"], name='New Recovered'),
    go.Bar(x=df_bd["Date"], y=df_bd["New Confirmed"], name='New Confirmed')
])
fig.update_layout(title_text='Comparision with New Confirmed and New Recovered', xaxis_tickfont_size=12, yaxis=dict(title='Cases'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_bd['Date'], y = df_bd['New Recovered'], mode='lines+markers', name='New Recovered',line=dict(color='Blue', width=2)))
fig.add_trace(go.Scatter(x=df_bd['Date'], y = df_bd['New Confirmed'], mode='lines+markers', name='New Confirmed',line=dict(color='Red', width=2)))
fig.update_layout(title='Comparision with New Confirmed and New Recovered', xaxis_tickfont_size=12, yaxis=dict(title='Cases'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```

```
df1_data= pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv').query('location == "Bangladesh" and total_cases > 0')
df1 = df1_data.copy().fillna(0)
df1.head(10)
```

```
df1 = df1.rename(columns={'date': 'Date', 'location': 'Country', 'new_tests': 'New Tests', 'total_tests': 'Tests'})
# dff = df1[['Date', 'Country']]
df1 = df1[['Date', 'New Tests', 'Tests']]
df1
```

```
# df = pd.concat([df_bd, df1]).drop_duplicates('Date').reset_index(drop=True).fillna(0)
df = pd.merge(df_bd, df1, left_on='Date', right_on='Date').reset_index(drop=True).fillna(0)
df = df[['Date', 'Country', 'New Confirmed', 'Confirmed', 'New Recovered', 'Recovered', 'New Deaths', 'Deaths', 'New Tests', 'Tests']]
# df.to_csv(r'Bangladesh-cases.csv', index = False) # download
df
```
