```
# importing the required libraries
import pandas as pd
import numpy as np

# Visualisation libraries
import plotly.express as px
import plotly.graph_objects as go
```

```
df_data= pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
df = df_data.copy()
# df.sort_values(by=['salary'], ascending=False, inplace=True) # sorting
# df.style.background_gradient(cmap='Greens')
# df['sl'] = list(range(len(df.index)))
df.head(10)
```

```
df_bd = df.query('location == "Bangladesh"')
# df_bd.to_csv(r'Bangladesh-cases.csv', index = False) # download
df_bd
```

```
fig = px.bar(df_bd, x="date", y="new_cases", barmode='group', height=600)
fig.update_layout(title_text='Coronavirus Cases in Bangladesh', plot_bgcolor='rgb(230, 230, 230)')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_bd['date'], y = df_bd['new_cases'], mode='lines+markers',name='Total Cases'))
fig.update_layout(title_text='Coronavirus Cases in Bangladesh', plot_bgcolor='rgb(230, 230, 230)')
fig.show()
```

```
# Past 14 days
fig = px.bar(df_bd.tail(14), x="date", y="new_cases", barmode='group', height=600)
fig.update_layout(title_text='Coronavirus Cases in Bangladesh Past 14 days', plot_bgcolor='rgb(230, 230, 230)')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_bd['date'].tail(14), y = df_bd['new_cases'].tail(14), mode='lines+markers',name='Total Cases'))
fig.update_layout(title_text='Coronavirus Cases in Bangladesh Past 14 days', plot_bgcolor='rgb(230, 230, 230)')
fig.show()
```

```
df_pak = df.query('location == "Pakistan"')
df_pak
```

```
# Past 14 days
fig = px.bar(df_pak.tail(14), x="date", y="new_cases", barmode='group', height=600)
fig.update_layout(title_text='Past 14 days', plot_bgcolor='rgb(230, 230, 230)')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_pak['date'].tail(14), y = df_pak['new_cases'].tail(14), mode='lines+markers',name='Total Cases'))
fig.update_layout(title_text='Past 14 days', plot_bgcolor='rgb(230, 230, 230)')
fig.show()
```

```
# Comparision with Bangladesh and Pakistan
fig = go.Figure(data=[
    go.Bar(x=df_bd["date"], y=df_bd["new_cases"], name='Bangladesh'),
    go.Bar(x=df_pak["date"], y=df_pak["new_cases"], name='Pakistan')
])
fig.update_layout(title_text='Comparision with Bangladesh and Pakistan', xaxis_tickfont_size=12, yaxis=dict(title='Cases'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_bd['date'], y = df_bd['new_cases'], mode='lines+markers', name='Bangladesh',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df_pak['date'], y = df_pak['new_cases'], mode='lines+markers', name='Pakistan',line=dict(color='Red', width=2)))
fig.update_layout(title='Comparision with Bangladesh and Pakistan', xaxis_tickfont_size=12, yaxis=dict(title='Cases'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```

Focasting

```
from fbprophet import Prophet
```

```
cases.columns = ['ds','y']
#cases['ds'] = cases['ds'].dt.date
cases['ds'] = pd.to_datetime(cases['ds'])
cases.tail()
```

```
m = Prophet(interval_width=0.95)
m.fit(cases)
future = m.make_future_dataframe(periods=7)
future.tail()
```

```
#predicting the future with date, and upper and lower limit of y value
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```
from fbprophet.plot import plot_plotly, plot_components_plotly
# cases_forecast_plot = m.plot(forecast)
plot_plotly(m, forecast)
```

```
plot_components_plotly(m, forecast)
```

```
# deaths
deaths.columns = ['ds','y']
deaths['ds'] = pd.to_datetime(deaths['ds'])

m = Prophet(interval_width=0.95)
m.fit(deaths)
future = m.make_future_dataframe(periods=7)
future.tail()
```
