```
tests = df.groupby('date').sum()['total_tests'].reset_index().tail(180)
cases = df.groupby('date').sum()['new_cases'].reset_index().tail(180)
deaths = df.groupby('date').sum()['new_deaths'].reset_index().tail(180)
print(cases)
```

```
fig = go.Figure()
#Plotting datewise confirmed cases
fig.add_trace(go.Scatter(x=tests['date'], y=tests['total_tests'], mode='lines+markers', name='New Tests',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=cases['date'], y=cases['new_cases'], mode='lines+markers', name='New Cases', line=dict(color='Red', width=2)))
fig.add_trace(go.Scatter(x=deaths['date'], y=deaths['new_deaths'], mode='lines+markers', name='New Deaths', line=dict(color='Green', width=2)))
fig.update_layout(title='Bangladesh Last 20 Days NCOVID-19 New Tests, Cases & Deaths', xaxis_tickfont_size=14,yaxis=dict(title='Number of Cases'))

fig.show()
```

```
from fbprophet import Prophet
```

```
cases.columns = ['ds','y']
#cases['ds'] = cases['ds'].dt.date
cases['ds'] = pd.to_datetime(cases['ds'])
```

```
cases.tail()
```

```
m = Prophet(interval_width=0.95)
m.fit(cases)
future = m.make_future_dataframe(periods=7)
future.tail()
```

```
#predicting the future with date, and upper and lower limit of y value
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```
cases_forecast_plot = m.plot(forecast)
```

```
cases_forecast_plot =m.plot_components(forecast)
```

```
# Another way to plot
from fbprophet.plot import plot_plotly, plot_components_plotly
# cases_forecast_plot = m.plot(forecast)
plot_plotly(m, forecast)
```

```
plot_components_plotly(m, forecast)
```
