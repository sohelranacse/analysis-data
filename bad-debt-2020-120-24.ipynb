```
# importing the required libraries
import pandas as pd
import numpy as np

# Visualisation libraries
import plotly.express as px
import plotly.graph_objects as go
```

```
df_data= pd.read_excel('bad-debt-2020-12-24.xlsx')
df = df_data.copy()
df.sort_values(by=['Sales Center'], inplace=True) # sorting by sales center
df.style.background_gradient(cmap='Reds')
```

```
# Sales Comparision
df_sales = format(df['Starting Bad debt'].sum()+df['Sales'].sum(), ',.2f')
df_sales1 = format(df['Collection'].sum(), ',.2f')
title_text = "Bad Debt Report Starting Bad Debt amount("+ str(df_sales)+"tk. ) & Collection("+ str(df_sales1)+"tk.)"

fig = go.Figure(data=[
    go.Bar(x=df["Sales Center"], y=df["Starting Bad debt"]+df["Sales"], name='Sales'),
    go.Bar(x=df["Sales Center"], y=df["Collection"], name='Collection')
])
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title_text=title_text, xaxis_tickfont_size=12, yaxis=dict(title='Sales Amount'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Starting Bad debt']+df["Sales"], mode='lines+markers', name='Sales',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Collection'], mode='lines+markers', name='Collection',line=dict(color='Green', width=2)))
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title=title_text, xaxis_tickfont_size=12, yaxis=dict(title='Amount'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```

```
# Collection PIE Chart
fig = px.pie(df, names='Sales Center', values='Collection', title='Collection of Sales Center')
fig.update_traces(textposition='inside', textfont_size=12)
fig.show()
```
