```
# importing the required libraries
import pandas as pd
import numpy as np

# Visualisation libraries
import plotly.express as px
import plotly.graph_objects as go
```

```
df_data= pd.read_excel('winter-offer-2019-2020.xlsx')
df = df_data.copy()
df.sort_values(by=['Sales Center'], inplace=True) # sorting by sales center
df.style.background_gradient(cmap='Reds')
```

```
# Booking, Sold & Collection Report
fig = go.Figure(data=[
    go.Bar(x=df["Sales Center"], y=df["Booking Amount"], name='Booking Amount'),
    go.Bar(x=df["Sales Center"], y=df["Sold Amount"], name='Sold Amount'),
    go.Bar(x=df["Sales Center"], y=df["Collection Amount"], name='Collection Amount')
])
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title_text='Booking, Sold & Collection Report', xaxis_tickfont_size=12, yaxis=dict(title='Booking, Sold & Collection Amount'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Booking Amount'], mode='lines+markers', name='Booking Amount',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Sold Amount'], mode='lines+markers', name='Sold Amount',line=dict(color='Red', width=2)))
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Collection Amount'], mode='lines+markers', name='Collection Amount',line=dict(color='Green', width=2)))
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title='Sales Report', xaxis_tickfont_size=12, yaxis=dict(title='Booking, Sold & Collection Amount'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```

```
df_data= pd.read_excel('winter-offer-2020-2021.xlsx')
df1 = df_data.copy()
df1.sort_values(by=['Sales Center'], inplace=True) # sorting by sales center
df1.style.background_gradient(cmap='Reds')
```

```
# Comparision Booking Report
fig = go.Figure(data=[
    go.Bar(x=df["Sales Center"], y=df["Booking Amount"], name='Booking Amount 2019'),
    go.Bar(x=df1["Sales Center"], y=df1["Booking Amount"], name='Booking Amount 2020'),
])
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title_text='Booking Comparision with 2019 & 2020', xaxis_tickfont_size=12, yaxis=dict(title='Booking Amount'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Booking Amount'], mode='lines+markers', name='Booking Amount 2019',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df1['Sales Center'], y = df1['Booking Amount'], mode='lines+markers', name='Booking Amount 2020',line=dict(color='Green', width=2)))
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title='Booking Comparision with 2019 & 2020', xaxis_tickfont_size=12, yaxis=dict(title='Booking Amount'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```

```
# Comparision Collection Report
fig = go.Figure(data=[
    go.Bar(x=df["Sales Center"], y=df["Collection Amount"], name='Collection Amount 2019'),
    go.Bar(x=df1["Sales Center"], y=df1["Collection Amount"], name='Collection Amount 2020'),
])
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title_text='Collection Comparision with 2019 & 2020', xaxis_tickfont_size=12, yaxis=dict(title='Collection Amount'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Collection Amount'], mode='lines+markers', name='Collection Amount 2019',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df1['Sales Center'], y = df1['Collection Amount'], mode='lines+markers', name='Collection Amount 2020',line=dict(color='Green', width=2)))
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title='Collection Comparision with 2019 & 2020', xaxis_tickfont_size=12, yaxis=dict(title='Collection Amount'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```

```
# Booking, Sold & Collection Comparision with previous year(2019-2020 And 2020-2021) Report
fig = go.Figure(data=[
    go.Bar(x=df["Sales Center"], y=df["Booking Amount"]-df1["Booking Amount"], name='Booking Amount'),
    go.Bar(x=df["Sales Center"], y=df["Sold Amount"]-df1["Sold Amount"], name='Sold Amount'),
    go.Bar(x=df["Sales Center"], y=df["Collection Amount"]-df1["Collection Amount"], name='Collection Amount')
])
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title_text='Booking, Sold & Collection Report', xaxis_tickfont_size=12, yaxis=dict(title='Booking, Sold & Collection Amount'), plot_bgcolor='rgb(230, 230, 230)')
fig.update_layout(barmode='group')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Booking Amount']-df1['Booking Amount'], mode='lines+markers', name='Booking Amount',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Sold Amount']-df1['Sold Amount'], mode='lines+markers', name='Sold Amount',line=dict(color='Red', width=2)))
fig.add_trace(go.Scatter(x=df['Sales Center'], y = df['Collection Amount']-df1['Collection Amount'], mode='lines+markers', name='Collection Amount',line=dict(color='Green', width=2)))
fig.update_traces(hovertemplate='%{x}<br>Tk. %{y:.0f}')
fig.update_layout(title='Sales Report', xaxis_tickfont_size=12, yaxis=dict(title='Booking, Sold & Collection Amount'), plot_bgcolor='rgb(230, 230, 230)', height=600)
fig.show()
```
